```python
# 5.4.1 대푯값 계산
# 평균, 중앙값, 최빈값 계산
#

import numpy as np
from statistics import mode
from scipy import stats

# 예제 데이터 집합: 학생들의 시험 점수
exam_scores = [82, 90, 76, 92, 88, 79, 95, 87, 90, 76, 92, 90, 85, 78, 85]

# 평균 계산
mean_score = np.mean(exam_scores)

# 중앙값 계산
median_score = np.median(exam_scores)

# 최빈값 계산
try:
    mode_score = mode(exam_scores)
except:
    mode_score = stats.mode(exam_scores)[0][0]

print(f"평균 점수: {mean_score}")
print(f"중앙값 점수: {median_score}")
print(f"최빈값 점수: {mode_score}")
```

```python
import numpy as np
# 여기서 scipystats 대신 scipy.stats로 수정했습니다.
from scipy.stats import variation

# 예제 데이터 집합
data = [1, 2, 2, 3, 4, 5, 5, 5, 6]

# 범위 계산
data_range = np.max(data) - np.min(data)

# 분산 계산
variance = np.var(data, ddof=0) # 모집합단 분산
sample_variance = np.var(data, ddof=1) # 표본 분산

# 표준편차 계산
std_deviation = np.std(data, ddof=0) # 모집합단 표준편차
sample_std_deviation = np.std(data, ddof=1) # 표본 표준편차

# 변동계수 계산 (모집단 기준)
cv = variation(data) # scipy.stats.variation 사용

print(f"범위: {data_range}")
print(f"분산(모집단): {variance}")
print(f"분산(표본): {sample_variance}")
print(f"표준편차(모집단): {std_deviation}")
print(f"표준편차(표본): {sample_std_deviation}")
print(f"변동계수: {cv}")
```

```python
# 필요한 라이브러리 임포트
import random

# 시뮬레이션을 위한 변수 초기화
trials = 10000
even_rolls_count = 0

# 10,000번의 주사위 던지기 시뮬레이션
for _ in range(trials):
    roll = random.randint(1, 6) # 1부터 6까지 무작위 수 (주사위 눈금)
    if roll % 2 == 0: # 짝수 눈금 확인
        even_rolls_count += 1 # 짝수 눈금이 나오면 카운트 증가

# 짝수 눈금이 나올 확률 계산
probability_even = even_rolls_count / trials

# 결과 출력
print(probability_even)
```

```python
# 표본공간 S와 사건 A, B 정의
S = {1, 2, 3, 4, 5, 6}
A = {1, 2}
B = {1, 3, 5}

# 합집합: A ∪ B
union_AB = A.union(B)

# 교집합: A ∩ B
intersection_AB = A.intersection(B)

# 여집합: A' (A의 여집합)
# 표본 공간 S에서 A의 원소를 제외
complement_A = S.difference(A)

# 여집합: B' (B의 여집합)
# 표본 공간 S에서 B의 원소를 제외
complement_B = S.difference(B)

# 결과 출력: 합집합, 교집합, A의 여집합, B의 여집합 순서
print((union_AB, intersection_AB, complement_A, complement_B))
```

```python
# 각 색의 공의 개수
red_balls = 2
blue_balls = 3
green_balls = 5

# 특정색(빨간색) 공이 나올 경우의 수
favorable_outcomes = red_balls

# 상자 안의 모든 가능한 결과의 수 (전체 공의 개수)
total_outcomes = red_balls + blue_balls + green_balls

# 특정색(빨간색) 공이 나올 확률 계산
probability = favorable_outcomes / total_outcomes

# 결과 출력
print(probability)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 주사위 던지기 시뮬레이션 횟수 조정: 1,000회
n_throws_short = 1000

# 주사위 던지기 결과 생성 (1~6 사이의 무작위 정수, 1,000회)
throws_short = np.random.randint(1, 7, size=n_throws_short)

# 4가 나온 횟수를 기록할 배열 초기화
# 누적 횟수를 저장하기 위해 전체 던지기 횟수만큼의 크기로 만듭니다.
fours_count_short = np.zeros(n_throws_short)

# 각 던지기마다 4가 나온 횟수를 업데이트 (누적합 계산)
# 첫 번째 던지기(i=0)부터 시작하며, throws_short[i]가 4면 1, 아니면 0을 더합니다.
is_four = (throws_short == 4).astype(int)
fours_count_short = np.cumsum(is_four)

# 각 시점에서의 확률 계산: (누적 4의 횟수) / (누적 던지기 횟수)
probabilities_short = fours_count_short / np.arange(1, n_throws_short + 1)

# 결과 그래프로 표시
plt.figure(figsize=(10, 6))

# 시뮬레이션 확률 곡선
plt.plot(probabilities_short, label='Simulated Probability', color='blue')

# 이론적 확률 (1/6)을 나타내는 수평선
plt.axhline(y=1/6, color='r', linestyle='--', label='Theoretical Probability (1/6)')

plt.xlabel('Number of Throws')
plt.ylabel('Probability of Throwing a 4')
plt.title('Probability of Throwing a 4 Converging to 1/6 (1,000 Throws)')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 주머니 안의 공 정의
# 'R' 3개 (빨간색), 'B' 4개 (파란색) -> 총 7개의 공
bag = ['R'] * 3 + ['B'] * 4

# 시뮬레이션 횟수 (복원 추출 횟수)
n_picks = 10000

# 복원 추출 시뮬레이션: 10,000번 추출
picks = np.random.choice(bag, size=n_picks, replace=True)

# 각 뽑기마다 빨간색 공을 뽑았는지 확인 (True/False 배열) 후 누적합 계산
# (picks == 'R')은 True(1), False(0)로 간주되어 누적 빨간색 공 횟수를 기록
red_count_cumulative = np.cumsum(picks == 'R')

# 각 시점에서의 실험적 확률 계산: (누적 빨간색 횟수) / (누적 뽑기 횟수)
probabilities_red_cumulative = red_count_cumulative / np.arange(1, n_picks + 1)

# 결과 그래프로 표시
plt.figure(figsize=(10, 6))

# 시뮬레이션 확률 곡선
plt.plot(probabilities_red_cumulative, label='Simulated Probability of Red', color='blue')

# 이론적 확률 (3/7)을 나타내는 수평선
# 이론적 확률: P(R) = (빨간 공 개수) / (총 공 개수) = 3 / 7
plt.axhline(y=3/7, color='r', linestyle='--', label='Theoretical Probability (3/7)')

plt.xlabel('Number of Picks')
plt.ylabel('Probability of Picking a Red Ball')
plt.title('Probability of Picking a Red Ball Converging to 3/7 (10,000 Picks)')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()
```
